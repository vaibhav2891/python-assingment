## Q1 Max and Min

```python
t=(1,5,2,8,3)
max_val=t[0]
min_val=t[0]
for i in t:
    if i>max_val: 
        max_val=i
    if i<min_val: 
        min_val=i
print('Max:',max_val,'Min:',min_val)
```

## Q2 Count occurrences

```python
t=(1,2,3,2,1,3,2)
freq={}
for x in t:
    freq[x]=freq.get(x,0)+1
print(freq)
```

## Q3 Join tuple strings

```python
t=('Hello','World','Python')
result=''
for word in t:
    result+=word
print(result)
```

## Q4 Sum of even numbers

```python
t=(1,2,3,4,5,6)
sum_even=0
for x in t:
    if x%2==0: 
        sum_even+=x
print('Sum of evens:',sum_even)
```
