## Student Class

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def grade(self):
        if self.marks>=80:
            return 'A'
        else:
            return 'B'

s=Student('vaibhav',85)
print(s.name,s.grade())
```

## Car Class

```python
class Car:
    def __init__(self,brand,year):
        self.brand=brand
        self.year=year
c=Car('Toyota',2020)
print(c.brand,c.year)
```
