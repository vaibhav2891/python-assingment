## Q1 Count vowels

```python
s=input('Enter string: ')
vowels='aeiouAEIOU'
count=0
for ch in s:
    if ch in vowels:
        count+=1
print('Vowels:',count)
```

## Q2 Palindrome

```python
s=input('Enter string: ')
rev=s[::-1]
if s==rev:
    print('Palindrome')
else:
    print('Not Palindrome')
```

## Q3 Reverse string

```python
s=input('Enter string: ')
rev=''
for ch in s:
    rev=ch+rev
print('Reverse:',rev)
```

## Q4 Character frequency

```python
s=input('Enter string: ')
freq={}
for ch in s:
    freq[ch]=freq.get(ch,0)+1
print(freq)
```

## Q5 Remove spaces

```python
s=input('Enter string: ')
print(s.replace(' ',''))
```
