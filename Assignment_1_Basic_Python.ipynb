## Q1 Average of list

```python
numbers_list=[10,20,30,40,50]
total_sum=0
for n in numbers_list:
    total_sum+=n
print('Average:', total_sum/len(numbers_list))
```

## Q2 Duplicate check

```python
nums=[32,10,45,90,45,6]
if len(nums)!=len(set(nums)):
    print('DUPLICATES')
else:
    print('ALL UNIQUE')
```

## Q3 Range check

```python
n=int(input('Enter number: '))
if 1<=n<=50:
    print('Ok')
else:
    print('Out of range')
```

## Q4 Sum of digits

```python
n=int(input('Enter number: '))
s=0
while n>0:
    s+=n%10
    n//=10
print('Sum:',s)
```

## Q5 Fibonacci

```python
n=int(input('Terms: '))
a,b=0,1
print(a,b,end=' ')
for i in range(n-2):
    c=a+b
    print(c,end=' ')
    a,b=b,c
```

## Q6 Prime till n

```python
n=int(input('Enter n: '))
for x in range(2,n+1):
    prime=True
    for i in range(2,x):
        if x%i==0:
            prime=False
    if prime:
        print(x)
```

## Q7 Pattern

```python
for i in range(4,0,-1):
    for j in range(1,i+1):
        print(j,end=' ')
    print()
```

## Q8 Reverse number

```python
n=int(input('Enter number: '))
r=0
while n>0:
    r=r*10+n%10
    n//=10
print('Reverse:',r)
```

## Q9 Sequential Search

```python
arr=[10,20,30,40,50]
t=int(input('Search: '))
found=False
for i in range(len(arr)):
    if arr[i]==t:
        print('Found at index',i)
        found=True
        break
if not found:
    print('Not found')
```
