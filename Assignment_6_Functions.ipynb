## Q1 Max of three numbers

```python
def max_three(a,b,c):
    return max(a,b,c)
print(max_three(10,28,5))
```

## Q2 Prime check

```python
def is_prime(n):
    if n<2: return False
    for i in range(2,n):
        if n%i==0: return False
    return True
print(is_prime(7))
```

## Q3 Factorial

```python
def fact(n):
    f=1
    for i in range(1,n+1): 
        f*=i
    return f
print(fact(5))
```

## Q4 Unique list

```python
def unique(lst):
    return list(set(lst))
print(unique([1,2,2,3]))
```

## Q5 Sum of evens

```python
def sum_even(lst):
    s=0
    for x in lst:
        if x%2==0:
            s+=x
    return s
print(sum_even([1,2,3,4]))
```
